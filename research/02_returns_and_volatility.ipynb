Features we'll be creating :-
1. Daily Log Returns
2. Absolute Returns
3. Rolling Volatility
4. volumne change
5. “I engineered return- and volatility-based features using rolling statistics to capture short-, medium-, and long-term market risk behavior.”

```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Data Science Learning\Projects\Volatility Regime Detection in Stock Returns\data\Processed_data\nifty_processed.csv")
df.head()
```

```python
#Compute Daily Log Returns
df["log_returns"] = np.log(df["close"]/df["close"].shift(1))
```

```python
df.head(5)
```

```python
#Compute Absolute Returns
df["abs_returns"] = df["log_returns"].abs()
```

```python
#Rolling Volatility
df["vol_7"] = df["log_returns"].rolling(window=7).std()
df["vol_14"] = df["log_returns"].rolling(window=14).std()
df["vol_30"] = df["log_returns"].rolling(window=30).std()
```

```python
df.head()
```

```python
#Volumne Change
df["volumn_change"] = df["turnover"].pct_change()
```

```python
df.head()
```

```python
#Drop Nans from rolling windows
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.to_csv(r"C:\Data Science Learning\Projects\Volatility Regime Detection in Stock Returns\data\03_Feature_Extraction\nifty_features.csv",index=False)
```
