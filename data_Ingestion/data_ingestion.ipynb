```python
import pandas as pd
import os
from glob import glob
```

```python
os.getcwd()
```

```python
raw_data_path=r"C:\Data Science Learning\Projects\Volatility Regime Detection in Stock Returns\data\Nifty_historical_index_data"
```

```python
csv_files = glob(os.path.join(raw_data_path,"*.csv"))
print(f"Total files found{len(csv_files)}")
```

```python
df_list = []

for file in csv_files:
    df=pd.read_csv(file)
    df_list.append(df)

raw_df = pd.concat(df_list,ignore_index=True)

raw_df.head()
```

```python
raw_df.rename(columns={"Turnover (₹ Cr)":"Turnover"},inplace=True)
```

```python
raw_df.columns = raw_df.columns.str.strip().str.lower()
raw_df.columns
```

```python
required_cols = ["date","close"]

if "volume" in raw_df.columns:
    required_cols.append("volumn")
elif "turnover" in raw_df.columns:
    required_cols.append("turnover")

print(required_cols)
```

```python
df = raw_df[required_cols].copy()
```

```python
df.head()
```

```python
df["date"]=pd.to_datetime(df["date"],dayfirst=True,errors="coerce")
```

```python
df.head()
```

```python
#Cleaning Numerical Columns
for col in ["close","volunme"]:
    if col in df.columns:
        df[col] = (
            df[col].astype(str).str.replace(",","",regex=True).astype(float)
        )
```

```python
#Sorting values according to date
df = df.sort_values("date",ascending=True).reset_index(drop=True)
```

```python
#Handling Duplicate values
df.drop_duplicates(subset="date",inplace=True)
df.dropna(subset=["date","close"],inplace=True)
```

```python
print("Start Date:",df["date"].min())
print("End Date:",df["date"].max())
print("Total Row",df.shape[0])
df.head()
```

```python
PROCESSED_PATH = r"C:\Data Science Learning\Projects\Volatility Regime Detection in Stock Returns\data\02_Processed_data"
os.makedirs(PROCESSED_PATH,exist_ok=True)
df.to_csv(f"{PROCESSED_PATH}/nifty_processed.csv",index=False)
```
